```python
import os
import shutil

class Working_With_Files:
    
    @staticmethod
    def New(path, file_name, folder_name="", data=" "):
        
        if os.path.exists(file_name) or os.path.exists(folder_name):
            
            return False
        
        if folder_name != "":
        
            folder_name = path + folder_name
            os.mkdir(folder_name)
            file_name = folder_name + '/' + file_name
        
        else:
            
            if path[len(path) - 1] == '/':
                file_name = path + file_name
            else:    
                file_name = path + '/' + file_name
            
            
        with open(file_name,'w', encoding='utf-8') as f:
            f.write(data)  

        return True
        
    
    @staticmethod
    def Delete(path):
    
        if os.path.exists(path):
            if os.path.isfile(path):
                os.remove(path)
            else:
                os.rmdir(path)
            return True
        else:
            return False
        
    
    @staticmethod
    def Moving(old_path, new_path):
                
        if os.path.exists(old_path) and os.path.exists(new_path):
            os.replace(old_path, new_path)
            return True
        else:
            return False

    @staticmethod
    def Copy(source_path, dest_path):
        if os.path.exists(source_path) and os.path.exists(dest_path):
            shutil.copytree(source_path, dest_path)
            return True
        else:
            return False
   

print("Добро пожаловать!")    
print("1. Создать файл или папку:")
print("2. Удалить файл или папку")
print("3. Переместить файл или папку")
print("4. Копировать файли или папку")
print("5. Выход")

while(True):
    choice = input("Выберите необходимое действие: ")

    match choice:
        case '1':
            
            flag = False
            path = input("Введите путь: ")
            print()
            folder_name = input("Нужно создать папку: Да, Нет? ").lower()
            
            if folder_name == "да":
                folder_name = input("Введите название папки: ")
                flag = True
            name_file = input("Введите название файла: ")
            
            if (flag):
                if Working_With_Files.New(path, name_file, folder_name):
                    print("Файл и папка созданы!")
                    break
                       
            elif Working_With_Files.New(path, name_file):
                print("Инициализация успешна!")
                break
            
            print("Данне не инициализированны или уже существуют!!!")

        case '2':
            
            path = input("Введите путь и название файла или папки: ")
            
            if Working_With_Files.Delete(path):
                print("Данные удалены!!!")
            else:
                print("Данные не существуют!!!")               

        case '3':
            
            path_1 = input("Введите путь и название файла или папки для перемещения: ")
            path_2 = input("Введите путь для переноса данных: ")
            
            if Working_With_Files.Moving(path_1, path_2):
                print("Данные перенесены!")
            else:
                print("Путь не найден")
        
        case '4':
            
            path_copy_1 = input("Введите путь и название файла или папки для копирования: ")
            path_copy_2 = input("Введите путь для копирования данных: ")
            
            if Working_With_Files.Copy(path_copy_1, path_copy_2):
                print("Данные скопированны!")
            else:
                print("Путь не найден")
            
        case '5':
                break
```
